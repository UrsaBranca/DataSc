```python
all_teams_matches = (
    match_data[
      (match_data['season'] == '2015/2016')
    ]
      .loc[:, ('match_api_id', 'home_team_api_id', 'away_team_api_id', 'home_team_goal', 'away_team_goal')]
)

all_teams_matches.head()
```

```python
all_teams_matches['home_win'] = all_teams_matches.home_team_goal > all_teams_matches.away_team_goal
all_teams_matches['away_win'] = all_teams_matches.away_team_goal > all_teams_matches.home_team_goal
all_teams_matches
```

```python
home_wins = (
    all_teams_matches
      .groupby('home_team_api_id')
      .agg({'match_api_id': 'count', 'home_win' : 'sum'})
      .rename(columns = {'home_win':'wins', 'match_api_id': 'matches'})
)

home_wins
```

```python
away_wins = (
    all_teams_matches
      .groupby('away_team_api_id')
      .agg({'match_api_id': 'count', 'away_win' : 'sum'})
      .rename(columns = {'away_win':'wins', 'match_api_id': 'matches'})
)

away_wins
```

```python
df = (
    home_wins
      .add(away_wins, fill_value=0)
      .rename_axis('team_api_id', axis='rows')
)

df['win_rate'] = df.wins / df.matches
df
```

**Task 15 (2 points).** Determine the team with the maximum days' gap between matches in **England Premier League 2010/2011 season**. Return number of days in that gap. <br>
**Note**: a *gap* means the number of days between two consecutive matches of the same team.

```python
epl_id = league_data[league_data['name'] == 'England Premier League']['country_id'].item()
epl_id
```

```python
epl_matches = (
    match_data[
      (match_data['season'] == '2010/2011') & (match_data['country_id'] == epl_id)
    ]
      .loc[:, ('match_api_id', 'date', 'home_team_api_id', 'away_team_api_id')]
)
                    
epl_matches['date'] = pd.to_datetime(epl_matches['date'])
epl_matches
```

```python
melt_epl_matches = pd.melt(
    epl_matches, 
    id_vars=['match_api_id', 'date'], 
    value_vars=('home_team_api_id','away_team_api_id'), 
    var_name='type_of_play', 
    value_name='team_api_id', 
    col_level=None
)

melt_epl_matches
```

```python
# var 1: –∫—Ä–∏–≤–æ –∏ –ø–µ—á–∞–ª—å–Ω–æ - —á–µ—Ä–µ–∑ –¥–≤–æ–π–Ω—É—é —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫—É + –¥–æ–ø—Å—Ç–æ–ª–±—Ü—ã:

# melt_epl_matches = melt_epl_matches.sort_values(by=['team_api_id', 'date'])
# melt_epl_matches['diff'] = pd.to_datetime(melt_epl_matches['date']).diff().dt.days

# melt_epl_matches.sort_values(by=['team_api_id', 'date']).head(40)
```

```python
# var 2: –ø–æ–≤–µ—Å–µ–ª–µ–µ, —á–µ—Ä–µ–∑ –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫—É

players_ids = np.sort(melt_epl_matches['team_api_id'].unique())
players_ids
```

```python
g = (
    melt_epl_matches
      .sort_values(by=['date'])
      .groupby('team_api_id')
      .apply(lambda x: x['date'].diff().dt.days)
)

max_gap = [0, 0]
for id in players_ids:
    if g[id].max() > max_gap[1]:
        max_gap[1] = g[id].max()
        max_gap[0] = id
print(max_gap)

# –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ —Å–æ —Å–ª–æ–∂–Ω—ã–º–∏ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è–º–∏ –ø–æ –∫–æ–ª–æ–Ω–∫–∞–º
# df_price = df.groupby(by=['price_levels'],as_index=False)[['CRIM','LSTAT']].apply(lambda v: v.max()-v.min())
# df_price 
```
